Transformando em geodataframe

```python
list(zip(df['Longitude'],df['Latitude']))
```

```python
from shapely import Point
```

```python
pontos_geometricos = [Point(x) for x in zip(df['Longitude'], df['Latitude'])]
pontos_geometricos
```

```python
crs={'proj':'latlong','ellps':'WGS84','datum':'WGS84','no_defs':True}

geo_dados=gpd.GeoDataFrame(df,crs=crs,geometry=pontos_geometricos)
```

```python
geo_dados.to_file('DATASET.shp')
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(12, 8))


geo_dados.plot(ax=ax, alpha=0.2, color='green', edgecolor='black')

plt.show()
```

Analisar a distância dos imóveis ao metrô e analisar se influencia no preço

```python
geo_dados.iloc[0].geometry
```

```python
distancias = metro.distance(geo_dados.iloc[0].geometry)
distancias_km = distancias / 1000
distancias_km
```

```python
distancias_km.min()
```

```python
geo_dados['Distancia_metro']=geo_dados['geometry'].apply(lambda x: metro.distance(x).min() )
```

```python
correlacao = geo_dados.select_dtypes(include=['number']).corr()
```

```python
import seaborn as sns

sns.heatmap(correlacao,annot=True,cmap='RdYlBu',linewidth=0.5)
```
